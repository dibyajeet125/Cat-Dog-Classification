```python
# Importing tensorflow and keras libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models

# defining the model and setup 3 convolutional layers, 3 maxpooling layers and then 2 dense layers in the end.
# setup will be like this
# conv2d-> maxpool2d, conv2d->maxpool2d, conv2d->maxpool2d, flatten layer, dense layer, dense layer

model = models.Sequential()
# add layers here
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
# Print the summary
model.summary()
```

```python
# Import optimizers and then compile your model, use loss='binary_crossentropy' and learning_rate=1e-4 with metrics=['acc']

from tensorflow.keras import optimizers

# compile your model
model.compile(loss = 'binary_crossentropy',
              optimizer = optimizers.RMSprop(learning_rate=1e-4),
             metrics = ['acc'])
```

**Feature extraction using pretrained VGG16**

```python
# Using pretrained network ImageNet (Method = Feature extraction)
#import VGG16 from tensorflow
from tensorflow.keras.applications import VGG16

# instantantiating VGG16 and pass include_top = False, weights='imagenet' and input shape = (150,150,3)

conv_base = VGG16(weights='imagenet',
                  include_top = False, input_shape = (150, 150, 3)) # code here
```

```python
# printing summary
conv_base.summary()
```

```python
import numpy as np
# location of main directories(base_dir, train_dir, validation_dir, test_dir )
base_dir = "/content/drive/MyDrive/Colab Datasets/cats_dogs_light/cats_vs_dogs"
train_dir = os.path.join(base_dir, 'train') # code here
validation_dir = os.path.join(base_dir, 'validation') # code here
test_dir = os.path.join(base_dir, 'test') # code here


# image datagen and batch size
datagen = ImageDataGenerator(rescale=1./255)
batch_size = 25

# Read code line by line and try to understand what it does
# defining feature_extraction function (feature extraction without data augmentation)
def feature_extraction(directory, sample):
  features = np.zeros(shape=(sample, 4, 4, 512)) # define an array of zeros with shape (sample, 4, 4, 512)
  labels = np.zeros(shape=(sample)) # define an array of  shape=(sample)
  generator = datagen.flow_from_directory(
      directory,
      target_size=(150, 150),
      batch_size = batch_size,
      class_mode = 'binary'
  )
  count = 0
  for image_batch, labels_batch in generator:
    batch_feature = conv_base.predict(image_batch) # predict using conv_base on image batch

    # Get the actual batch size
    actual_batch_size = batch_feature.shape[0]

    features[count * batch_size : count * batch_size + actual_batch_size] = batch_feature
    labels[count * batch_size : count * batch_size + actual_batch_size] = labels_batch
    count += 1
    if count * batch_size >= sample:
      break
  return features, labels
```

```python
# Now extract features for train, validation, test

train_features, train_labels = feature_extraction(train_dir, 10500) # code here
validation_features, validation_labels = feature_extraction(validation_dir, 10000) # code here
test_features, test_labels = feature_extraction(test_dir, 10000) # code here
```

```python
# reshape, train, validation, test into (no of samples, 4*4*512) features before passing into dense layers
train_features = np.reshape(train_features, (10500, 4*4*512)) # code here
validation_features = np.reshape(validation_features, (10000, 4*4*512)) # code here
test_features = np.reshape(test_features, (10000, 4*4*512)) # code here
```

```python
# define a new model and  add dense classfier first then dropout(0.5) and then final dense layer with activation sigmoid

model = models.Sequential()
# code here
model.add(layers.Dense(128, activation='relu', input_dim=4*4*512))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer = optimizers.RMSprop(learning_rate=1e-5),
              loss='binary_crossentropy',
              metrics = ['acc'])
```

```python
# fit the model with epochs = 10 and batch_size=20
vgg_history = model.fit(train_features, train_labels,
                        epochs=10, batch_size=20,
                        validation_data=(validation_features, validation_labels)) #code here
```

```python
# define acc, loss, val_acc, val_loss again from history.history method

acc = vgg_history.history['acc'] # code here
val_acc = vgg_history.history['val_acc'] # code here
loss = vgg_history.history['loss'] # code here
val_loss = vgg_history.history['val_loss'] # code here

# then plot epochs vs train accuracy and validation accuracy
epochs = range(1, len(acc) + 1)
# code here
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Val acc')
plt.title('Training and validation accuracy')
plt.legend()
plt.show()
```

```python
# plot epochs vs train loss and validation loss
# code here
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Val loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```
